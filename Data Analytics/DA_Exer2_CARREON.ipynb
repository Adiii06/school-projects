### Question 3

```python
# Load the first 200 rows called 'dataset'
dataset = pd.read_excel("assignment-data.xlsx", sheet_name="online-retail", nrows=200)
```

#### Task 1

```python
# Data Cleaning
ncols = dataset.select_dtypes(include=['number'])
bfr = (ncols < 0).values.any()
clean_df = dataset[(ncols >= 0).all(axis=1)]
aft = (clean_df.select_dtypes(include=['number']) < 0).values.any()

print(f"Before cleaning: any negatives?: \033[1m{bfr}\033[0m")
print(f"After cleaning: any negatives?: \033[1m{aft}\033[0m")
```

#### Task 2

```python
# Find Customer with most number of transactions
transcount = clean_df.groupby('CustomerID').size()
mt_customer = transcount.idxmax()
mt_count = transcount.max()

print(f"Hurray! The most number of transactions is done by Customer ID: \033[1m{mt_customer}\033[0m. He has made \033[1m{mt_count}\033[0m transactions. He will be rewarded with our gift hamper worth 1000 pesos.")
```

#### Task 3

```python
# Find Customer with highest amount of total purchase amount
clean2 = clean_df.copy()
clean2['TotalAmount'] = clean2['UnitPrice'] * clean2['Quantity']
tpa = clean2.groupby('CustomerID')['TotalAmount'].sum()
bc = tpa.idxmax()
bca = tpa.max()

print(f"Customer ID: \033[1m{bc}\033[0m is the best customer since he has spent a lot on our business. His total purchase amount is \033[1m{bca:.2f}\033[0m pesos. 50% discount will be given on his next purchase.")
```

#### Task 4

```python
# Find the most and the least bought items
quanitems = clean_df.groupby('StockCode')['Quantity'].sum()
mbi = quanitems.idxmax()
mbq = quanitems.max()
lbi = quanitems.idxmin()
lbq = quanitems.min()

print(f"Most bought item is StockCode \033[1m{mbi}\033[0m with \033[1m{mbq}\033[0m in purchased quantity.")
print(f"Least bought item is StockCode \033[1m{lbi}\033[0m with \033[1m{lbq}\033[0m in purchased quantity overall. These two items can be bundled and provide 10% discount on this bundle to increase the sale of StockCode \033[1m{lbi}\033[0m.")
```

#### Task 5

```python
# Find the overall average purchase-amount among all transactions
trans_tot = clean2.groupby('InvoiceNo')['TotalAmount'].sum()
apa = trans_tot.mean()

print(f"Average purchase amount among all transactions = \033[1m{apa:.2f}.\033[0m")
```

#### Task 6

```python
# Create new data frame called 'new-dataset' and display last 10 rows
new_dataset = clean2.groupby('CustomerID').agg(
    TotalPurchaseAmount=('TotalAmount', 'sum'),
    NumberofTransactions=('InvoiceNo', 'count')
).reset_index()

new_dataset.index += 1
display(new_dataset.tail(10))
```

#### Task 7

```python
# Know the average purchase amount made by every customer
new_dataset['AveragePurchase'] = new_dataset['TotalPurchaseAmount'] / new_dataset['NumberofTransactions']
new_dataset.index -= 1

new_dataset.head(5)
```

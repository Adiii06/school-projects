#### 1. Poisson Distribution

<h3>$$P_x=\frac{\lambda^x}{x!}e^{-\lambda}$$</h3>

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Sets seed 
np.random.seed(42)

# Sets parameter
lpoisson = 6.0
ts_poisson = 6.5
numsim_poisson = 10000

# Generates the data
sd_poisson = poisson.rvs(lpoisson, size=numsim_poisson)

# Filters the data based on test statistic
fd_poisson = sd_poisson[sd_poisson >= ts_poisson]

# Plots the histogram
plt.figure(figsize=(10, 6))
plt.hist(sd_poisson, bins=np.arange(0, max(sd_poisson) + 1.5) - 0.5, alpha=0.7, label="The Data", color="#800080", edgecolor="black")
plt.axvline(ts_poisson, linestyle='dashdot', linewidth=2, label="Test Statistic", color="Gray")
plt.hist(fd_poisson, bins=np.arange(0, max(sd_poisson) + 1.5) - 0.5, alpha=0.7, label="Filtered Data", color="#ADD8E6", edgecolor="black")
plt.title("Poisson Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Shows Observed and Theoretical Value
ov_poisson = np.mean(sd_poisson)
tv_poisson = poisson.mean(lpoisson)

print("\033[1mPoisson Distribution: \033[0m")
print(f"Observed Value = {ov_poisson:.4f}")
print(f"Theoretical Value = {tv_poisson:.4f}")
```

#### 2. Hypergeometric Distribution

<h2>$$P_x=\frac{\binom{g}{x}\binom{b}{n-x}}{\binom{g+b}{n}}$$</h2>

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import hypergeom

# Sets seed
np.random.seed(42)

# Sets parameter
g_hypergeom, b_hypergeom, n_hypergeom = 20, 22, 6
ts_hypergeom = 8
numsim_hypergeom = 11000

# Generates the data
sd_hypergeom = hypergeom.rvs(g_hypergeom + b_hypergeom, g_hypergeom, n_hypergeom, size=numsim_hypergeom)

# Filters the data based on test statistic
fd_hypergeom = sd_hypergeom[sd_hypergeom >= ts_hypergeom]

# Plots the histogram
plt.figure(figsize=(10, 6))
plt.hist(sd_hypergeom, bins=np.arange(0, max(sd_hypergeom) + 1.5) - 0.5, alpha=0.7, label="The Data", color="#800080", edgecolor="black")
plt.axvline(ts_hypergeom, linestyle='dashdot', linewidth=2, label="Test Statistic", color="black")
plt.hist(fd_hypergeom, bins=np.arange(0, max(sd_hypergeom) + 1.5) - 0.5, alpha=0.7, label="Filtered Data", color="#ADD8E6", edgecolor="black")
plt.title("Hypergeometric Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Shows Observed and Theoretical Value
ov_hypergeom = np.mean(sd_hypergeom)
tv_hypergeom = hypergeom.mean(g_hypergeom + b_hypergeom, g_hypergeom, n_hypergeom)

print("\033[1mHypergeometric Distribution: \033[0m")
print(f"Observed Value = {ov_hypergeom:.4f}")
print(f"Theoretical Value = {tv_hypergeom:.4f}")
```

#### 3. Exponential Distribution

<h3>$$P_x=\lambda e^{-\lambda x}$$</h3>

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Sets seed
np.random.seed(42)

# Sets parameter
lExponential = 0.6
ts_exponential = 2.0
numsim_exponential = 12000

# Generates the data
sd_exponential = expon.rvs(scale=1/lExponential, size=numsim_exponential)

# Filters the data based on test statistic
fd_exponential = sd_exponential[sd_exponential >= ts_exponential]

# Plots the histogram
plt.figure(figsize=(10, 6))
plt.hist(sd_exponential, bins=np.arange(0, max(sd_exponential) + 1.5) - 0.5, alpha=0.7, label="The Data", color="#800080", edgecolor="black")
plt.axvline(ts_exponential, color='black', linestyle='dotted', linewidth=2, label="Test Statistic")
plt.hist(fd_exponential, bins=np.arange(0, max(sd_exponential) + 1.5) - 0.5, alpha=0.7, label="Filtered Data", color="#ADD8E6", edgecolor="black")
plt.title("Exponential Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Shows Observed and Theoretical Value
ov_exponential = np.mean(sd_exponential)
tv_exponential = expon.mean(scale=1/lExponential)

print("\033[1mExponential Distribution: \033[0m")
print(f"Observed Value = {ov_exponential:.4f}")
print(f"Theoretical Value = {tv_exponential:.4f}")
```

#### 4. Chi-Squared Distribution

<h3>$$P_x=\frac{x^{k/2-1}e^{-x/2}}{2^{k/2}\Gamma(k/2)}$$</h3>

```python
# Your code here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Sets seed
np.random.seed(42)

# Sets parameter
df_chi2 = 6.0
ts_chi2 = 8.0
numsim_chi2 = 13000

# Generates the data
sd_chi2 = chi2.rvs(df_chi2, size=numsim_chi2)

# Filters the data based on test statistic
fd_chi2 = sd_chi2[sd_chi2 >= ts_chi2]

# Plots the histogram
plt.figure(figsize=(10, 6))
plt.hist(sd_chi2, bins=np.arange(0, max(sd_chi2) + 1.5) - 0.5, alpha=0.7, label="The Data", color="#800080", edgecolor="black")
plt.axvline(ts_chi2, color='black', linestyle='dotted', linewidth=2, label="Test Statistic")
plt.hist(fd_chi2, bins=np.arange(0, max(sd_chi2) + 1.5) - 0.5, alpha=0.7, label="Filtered Data", color="#ADD8E6", edgecolor="black")
plt.title("Chi-Squared Distribution Simulation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Shows Observed and Theoretical Value
ov_chi2 = np.mean(sd_chi2)
tv_chi2 = chi2.mean(df_chi2)

print("\033[1mChi-Squared Distribution: \033[0m")
print(f"Observed Value = {ov_chi2:.4f}")
print(f"Theoretical Value = {tv_chi2:.4f}")
```

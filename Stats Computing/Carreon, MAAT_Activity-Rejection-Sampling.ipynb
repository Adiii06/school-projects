### Instruction:

Write the corresponding Python code that uses Numpy to show the given distribution of the given function using rejection sampling. 

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
   return np.exp(-x**2) * (np.cos(2*x)**2 * np.sin(5*x) + np.sin(x)**2 + 1)

def f2(x):
   return 2 - (np.arctan(x))**2

def rejection_sampling(func, x_min, x_max, y_min, y_max, ns = 10000):
   samples = []
   while len(samples) < ns:
       x = np.random.uniform(x_min, x_max)
       y = np.random.uniform(y_min, y_max)
       if y <= func(x):
           samples.append(x)
   return np.array(samples)

x1 = np.linspace(-2, 2, 1000)
x2 = np.linspace(-np.tan(np.sqrt(2)),np.tan(np.sqrt(2)), 1000)
y1 = f1(x1)
y2 = f2(x2)

s1 = rejection_sampling(f1, -2, 2, 0, 2)
s2 = rejection_sampling(f2, -np.tan(np.sqrt(2)),np.tan(np.sqrt(2)), 0, 2)

plt.figure(figsize=(12, 8))

plt.subplot(2, 1, 1)
plt.plot(x1, y1, label='f(x) = e^(-x^2)(cos^2(2x)sin(5x) + sin^2(x) + 1)', color='black')
plt.hist(s1, bins=50, density=True, alpha=0.7, label='Sampled Distribution', color='purple')
plt.title('Sampled Distribution using Rejection Sampling (Function 1)')
plt.legend()

# Function 2
plt.subplot(2, 1, 2)
plt.plot(x2, y2, label='f(x) = 2 - arctan2(x)', color='gray')
plt.hist(s2, bins=50, density=True, alpha=0.7, label='Sampled Distribution', color='#ADD8E6')
plt.title('Sampled Distribution using Rejection Sampling (Function 2)')
plt.legend()

plt.tight_layout()
plt.show()
```

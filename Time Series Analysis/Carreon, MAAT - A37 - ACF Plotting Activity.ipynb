```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

#### 1. Create a Random Walk Time Series from `numpy` with given specifications

The random walk should be:
- A set seed of `337` is given
- A random number generator `MT19937` should be used
- The data point starts at 100
- Simulate a coin flip: If it lands heads, add 1 to the previous data point, otherwise subtract 1 to the previous data point
- Append the result at the end of the series
- Repeat the process 500 times

Plot the given result and its ACF plot using `numpy`.

```python
# Write your code here. Add cells if necessary
# Random number generator MT19937 with seed 337
MT19937 = np.random.MT19937(337)
rng = np.random.default_rng(MT19937)

# Random walk series with the starting point
random_walk = [100]

# Generate the random walk series
for _ in range(500):
    # Flip a coin: 1 for heads (add 1), 0 for tails (subtract 1)
    coin_flip = rng.choice([-1, 1])
    # Update the random walk series
    random_walk.append(random_walk[-1] + coin_flip)

# Convert the list to a numpy array for easier manipulation
random_walk = np.array(random_walk)

# Plot the random walk time series
plt.figure(figsize=(10, 4))
plt.plot(random_walk, color='#5D3FD3')
plt.title('Random Walk Time Series')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

# Plot the ACF of the random walk series
plot_acf(random_walk, color='#6495ED')
plt.title('ACF of Random Walk Time Series')
plt.xlabel('Lags')
plt.show()
```

#### 2. Using the given data, decompose the given data and plot its ACF

The data is given in the variable `elec_consumption`

Perform the given tasks:
- Get the 1st differencing using the column `IPG2211A2N`
- Using the differenced value, show an ACF plot using `statsmodels`
- Get the 4th differencing using the column `IPG2211A2N`
- Using the newly differenced value, show an ACF plot using `statsmodels`

```python
elec_consumption = pd.read_csv('https://raw.githubusercontent.com/ejgao/Time-Series-Datasets/master/Electric_Production.csv')
elec_consumption.head()
```

```python
# Write your code here
# 1st differencing
elec_consumption_diff1 = np.diff(elec_consumption['IPG2211A2N'], n=1)

# Plot the ACF of the 1st differenced series
plot_acf(elec_consumption_diff1, color='#4169E1')
plt.title('ACF of 1st Differenced Series')
plt.xlabel('Lags')
plt.show()


# 4th differencing
elec_consumption_diff4 = np.diff(elec_consumption['IPG2211A2N'], n=4)

# Plot the ACF of the 4th differenced series
plot_acf(elec_consumption_diff4, color='#4169E1')
plt.title('ACF of 4th Differenced Series')
plt.xlabel('Lags')
plt.show()
```

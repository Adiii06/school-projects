#### What is Call of Duty?
Call of Duty (CoD) is a well-known first-person shooter (FPS) video game franchise published by Activision. It initially focused on World War II-themed games but has since expanded to various settings, including Cold War conflicts, futuristic environments, and present-day scenarios. Each release emphasizes immersive gameplay, competitive features, and multiple game modes that cater to both casual and professional gamers.

#### Overview of the Dataset
This dataset, COD_WLC.csv, captures detailed statistics from various professional matches in the Call of Duty World League Championships (2017-2019). The championships are highly competitive events featuring the world's best CoD players and teams, who compete in modes like Hardpoint, Search and Destroy, and Capture the Flag. Each row in this dataset represents an individual player's performance in a particular match and includes their team and player information, as well as performance metrics across different game modes, weapons, and in-game objectives.

The dataset includes around 18,000 rows and over 70 columns, detailing individual player statistics such as scores, kills, deaths, accuracy, and favorite strategies used in the matches. This data offers a granular look into player performance, team tactics, and game mode preferences, making it a valuable source for statistical analysis and performance insights.

#### Source of the Dataset
The data originates from Activision's Call of Duty World Championship series and is available as an open-source dataset on GitHub. It has been compiled from multiple sources into a single dataset, enabling a holistic view of professional-level CoD gameplay statistics.

#### Objective of the Dataset
With the growing eSports industry, datasets like COD_WLC.csv provide valuable insights for analyzing professional gaming trends. Professional players, analysts, and teams can use this data to identify patterns, improve strategies, and prepare for high-level competition. Additionally, this data can help the broader gaming community understand the dynamics of competitive play at the world championship level.

#### Major Insights Gained from the Dataset
1. **Performance Trends by Game Mode and Map**
Different game modes (e.g., Hardpoint, Search & Destroy) demand varied strategies and weapon preferences. By analyzing the data by mode and map, teams can tailor training programs to specific gameplay needs, and players can optimize load-outs based on typical success rates.

2. **Accuracy and Its Impact on Winning Outcomes**
Exploring the correlation between a player's shooting accuracy and the likelihood of winning offers insights into the importance of precision in different game scenarios. For instance, players with higher accuracy in Search & Destroy may contribute more to team success due to the high-stakes, elimination-based nature of the mode.

3. **Weapon Effectiveness and Player Preference**
Identifying which weapons yield the highest kill/death ratios per mode and map can inform players’ choices and strategies. Weapon recommendations can be made based on map type, allowing players to make data-driven load-out decisions.

4. **Player Consistency and Survival Analysis**
Metrics such as the average time per life and total time alive provide insight into player survival skills. Analyzing these trends could reveal defensive gameplay tactics that reduce deaths and increase player longevity in different match scenarios.

5. **Team Collaboration and Objective Handling**
The dataset contains rich details on team dynamics, such as assists, captures, and defends, which highlight collaborative play. Understanding how teams allocate roles in modes like Capture the Flag (CTF) and Hardpoint can inform optimal team configurations and improve in-game communication.

6. **Kill Streaks and Momentum Building**
The data provides information on players' kill streaks, which are essential for gaining scorestreak rewards. Analyzing streak patterns could reveal strategies for maintaining momentum, especially in game modes where consecutive kills yield high-impact rewards.

## Importing Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Preprocessing

```python
cod = pd.read_csv('COD_WLC.csv', low_memory=False)
```

```python
cod.head()
```

```python
cod.info()
```

```python
missing_values = cod.isnull().sum()
print("Missing values:\n", missing_values[missing_values > 0])
```

```python
cod = cod.dropna(thresh=len(cod) * 0.5, axis=1)
cod['accuracy (%)'] = cod['accuracy (%)'].str.replace('%', '').astype(float)
cod.fillna({
    'kills': cod['kills'].mean(),
    'deaths': cod['deaths'].median(),
    'accuracy (%)': cod['accuracy (%)'].mean()
}, inplace=True)
```

```python
cod['player score'].fillna(cod['player score'].mean(), inplace=True)
cod['player spm'].fillna(cod['player spm'].mean(), inplace=True)
cod['damage dealt'].fillna(cod['damage dealt'].median(), inplace=True)

cod['ekia'].fillna(cod['kills'] + cod['assists'], inplace=True)
cod['ekia'].fillna(cod['ekia'].mean(), inplace=True)  # Fill remaining with mean

cod['fave weapon'].fillna(cod['fave weapon'].mode()[0], inplace=True)

ctrl_columns = ['ctrl rounds', 'ctrl firstbloods', 'ctrl firstdeaths', 'ctrl captures']
cod[ctrl_columns] = cod[ctrl_columns].fillna(0)
```

```python
missing_values = cod.isnull().sum()
print("Remaining missing values:\n", missing_values[missing_values > 0])
```

```python
custom_palette = ["#D8BFD8", "#ADD8E6", "#FFB6C1", '#fbb4ae', '#b3cde3', '#ccebc5', '#decbe4', '#fed9a6', '#ffffcc']
```

## Data Visualization

### 1. Performance Trends by Game Mode and Map

```python
mode_map_kd = cod.groupby(['mode', 'map'])['k/d'].mean().unstack()
```

```python
pastel_cmap = sns.color_palette(custom_palette, as_cmap=True)
plt.figure(figsize=(12, 8))
sns.heatmap(mode_map_kd, annot=True, cmap=pastel_cmap, fmt=".2f")
plt.title("Kill/Death Ratios by Game Mode and Map")
plt.xlabel("Map")
plt.ylabel("Game Mode")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=cod, x='map', y='k/d', hue='mode', palette=custom_palette)
plt.title("Distribution of Kill/Death Ratios by Map and Game Mode")
plt.xlabel("Map")
plt.ylabel("K/D Ratio")
plt.xticks(rotation=45)
plt.legend(title='Game Mode')
plt.show()
```

### 2. Accuracy and Its Impact on Winning Outcomes

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='mode', y='accuracy (%)', hue='win?', data=cod, split=True, palette=custom_palette)
plt.title("Accuracy by Game Mode and Winning Outcome")
plt.xlabel("Game Mode")
plt.ylabel("Accuracy (%)")
plt.legend(title="Winning Outcome", loc="upper left")
plt.show()
```

```python
custom_palette = ["#decbe4", "#b3cde3"]

plt.figure(figsize=(12, 8))
sns.stripplot(x='mode', y='accuracy (%)', hue='win?', data=cod, palette=custom_palette, jitter=True, dodge=True)
plt.title("Accuracy by Game Mode and Winning Outcome")
plt.xlabel("Game Mode")
plt.ylabel("Accuracy (%)")
plt.legend(title="Winning Outcome", loc="upper left")
plt.show()
```

### 3. Weapon Effectiveness and Player Preference

```python
weapon_mode_kd = cod.groupby(['fave weapon', 'mode'])['k/d'].mean().unstack()

weapon_mode_kd.plot(kind='bar', stacked=True, figsize=(12, 8), color=custom_palette)
plt.title("Kill/Death Ratios by Favorite Weapon per Game Mode")
plt.xlabel("Weapon")
plt.ylabel("Kill/Death Ratio")
plt.legend(title="Game Mode")
plt.show()
```

```python
palette = [
    "#8A2BE2", "#6A5ACD", "#9370DB", "#BA55D3", "#FF69B4", "#FFB6C1", "#ADD8E6", "#B0E0E6"]

plt.figure(figsize=(12, 8))
for i, weapon in enumerate(weapon_mode_kd.index):
    plt.plot(weapon_mode_kd.columns, weapon_mode_kd.loc[weapon], marker='o', label=weapon, color=palette[i % len(palette)])

plt.title("Kill/Death Ratios by Favorite Weapon per Game Mode")
plt.xlabel("Game Mode")
plt.ylabel("Kill/Death Ratio")
plt.legend(title="Favorite Weapon", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.tight_layout()
plt.show()
```

### 4. Player Consistency and Survival Analysis

```python
cod['avg time per life (s)'] = pd.to_numeric(cod['avg time per life (s)'], errors='coerce')
cod.dropna(subset=['avg time per life (s)'], inplace=True)

plt.figure(figsize=(10, 6))
sns.boxplot(x='mode', y='avg time per life (s)', data=cod, palette=custom_palette)
plt.title("Average Time per Life by Game Mode")
plt.xlabel("Game Mode")
plt.ylabel("Time per Life (seconds)")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(cod['avg time per life (s)'], bins=15, kde=True, color='#BA55D3')
plt.title("Distribution of Average Time per Life")
plt.xlabel("Time per Life (seconds)")
plt.ylabel("Frequency")
plt.grid()
```

### 5. Team Collaboration and Objective Handling

```python
team_metrics = cod.groupby('mode')[['assists', 'hill captures', 'hill defends']].sum()

team_metrics.plot(kind='bar', stacked=True, figsize=(12, 8), color=custom_palette)
plt.title("Team Collaboration Metrics by Game Mode")
plt.xlabel("Game Mode")
plt.ylabel("Total Count")
plt.legend(title="Metric")
plt.show()
```

```python
team_metrics.plot(kind='line', marker='o', figsize=(12, 8), color=custom_palette)

plt.title("Team Collaboration Metrics by Game Mode")
plt.xlabel("Game Mode")
plt.ylabel("Total Count")
plt.legend(title="Metric")
plt.grid()
plt.show()
```

### 6. Kill Streaks and Momentum Building

```python
cod['time_segment'] = cod['duration (s)'] // 300
cod['kill_streak'] = (cod['kills'] > 0).astype(int).cumsum() - (cod['deaths'] > 0).astype(int).cumsum()
time_killstreaks = cod.groupby('time_segment')['kill_streak'].mean()

plt.figure(figsize=(10, 6))
time_killstreaks.plot(kind='line', marker='o', color='#7338A0')
plt.title("Average Kill Streak Progression Over Game Time")
plt.xlabel("Time Segment (5-minute intervals)")
plt.ylabel("Average Kill Streak")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
time_killstreaks.plot(kind='bar', color='#4B0082', alpha=0.7, edgecolor='black')
plt.title("Average Kill Streak Progression Over Game Time")
plt.xlabel("Time Segment (5-minute intervals)")
plt.ylabel("Average Kill Streak")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

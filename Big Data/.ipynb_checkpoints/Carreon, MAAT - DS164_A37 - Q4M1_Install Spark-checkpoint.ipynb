#### Install Spark on your machine

- Go to Apache Spark official sites download section and download the latest version of Spark [HERE](https://www.apache.org/dyn/closer.lua/spark/spark-3.5.1/spark-3.5.1-bin-hadoop3.tgz).
- Select **HTTP** and the download should start shortly after this.
- Go to your Disk C and create a folder named **Spark**.
- Create a folder inside named **bin**
- Visit this GitHub Repository [Here](https://github.com/steveloughran/winutils?source=post_page-----e16b7382624a--------------------------------) and find the folder below:<br>
![git_img](git.png)
- download **winutils.exe** and **hadoop.dll** and place them inside the **bin** folder you previously created

#### Setting Executable Environment Variables

```python
import os
import sys

os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
import findspark
findspark.init()
```

#### Create your Spark Session!

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession
        .builder
        .master('local[*]')
        .getOrCreate()
)
```

```python
spark
```

```python
from pyspark import Row

(spark.createDataFrame(
    [
        Row(a=1, b=2, c=3),
        Row(a=4, b=5, c=6)
    ]
)
    .show())
```

```python
import pandas as pd

pdf = pd.DataFrame({
    'a': [1, 4],
    'b': [2, 5],
    'c': [3, 6]
})
pdf
```

```python
(spark
    .createDataFrame(pdf)
    .show())
```

### Question 1

```python
def recursivecount(lst):
    if not lst: 
        raise ValueError("The given list is empty.")
    if len(lst) == 1:
        return 1
    
    return 1 + recursivecount(lst[1:])
```

```python
try:
    elist1 = []
    print("Count of empty list:", recursivecount(elist1))
except ValueError as e:
    print(e)

try:
    elist2 = [6, 10, 20, 26, 30]
    print("Count of non-empty list:", recursivecount(elist2))
except ValueError as e:
    print(e)
```

### Question 3

```python
class Parent:
    lists = [
        [5, 10, 15, 20, 25, 30],
        [6, 12, 18, 24, 36, 40],
        [50, 60, 70, 80, 90, 100],
        [13, 26, 39, 52, 65, 78]
    ]

class Child(Parent):
    @staticmethod
    def calculation():
        sorted_list = sorted(item for sublist in Parent.lists for item in sublist)
        lowest_product = sorted_list[0] * sorted_list[1]
        highest_product = sorted_list[-1] * sorted_list[-2]
        result = round(lowest_product / highest_product, 3)
        
        return f'({sorted_list[0]}*{sorted_list[1]}) / ({sorted_list[-1]}*{sorted_list[-2]}) = \033[1m{result}\033[0m'

print(Child.calculation())  
```

### Question 4

```python
class ListChecking:
    def __init__(self, superlist, sublist):
        self.superlist = superlist
        self.sublist = sublist

    def is_subset(self):
        return set(self.sublist).issubset(self.superlist)

superlist = [5, 6, 10, 20, 26, 30]

sublist1 = [6, 10, 20]
sublist2 = [2, 4, 6]
```

```python
for sublist in [sublist1, sublist2]:
    result = ListChecking(superlist, sublist).is_subset()
    message = "is" if result else "is not"
    print(f"The given sublist {sublist} \033[1m{message}\033[0m a subset of the superlist {superlist}.")
```
